```python
import numpy as np
import matplotlib.pyplot as plt

path = './dataset/images/puzzles/board_0.npy'

def load_array(image_path):
    return np.load(image_path, allow_pickle=True)

image_arr = load_array(path)

# Plot all 28x28 images in a single plot
fig, axes = plt.subplots(nrows=image_arr.shape[0], ncols=image_arr.shape[1], figsize=(10, 10))
for i, row in enumerate(image_arr):
    for j, img in enumerate(row):
        if img is not None:
            axes[i, j].imshow(img, cmap='gray')
            axes[i, j].set_title(f'({i}, {j})')
            axes[i, j].axis('off')
plt.tight_layout()
plt.show()
```
